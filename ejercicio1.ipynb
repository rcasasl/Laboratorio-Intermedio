```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ejercicio1.csv")
#print(df)
#df.plot()
x = [0.1,0.15,0.2,0.25,0.3,0.35,0.4]
y = [0.65,0.79,0.9,1.01,1.1,1.2,1.27]
plt.plot(x,y)
plt.show()

```

```python
f = np.sqrt(y)
sigmaf = (1/2*f)*0.01
#print(sigmaf)
plt.errorbar(x,f,yerr=sigmaf,xerr=0.01)
plt.show()
```

```python
T2 = np.square(x)
coefficients = np.polyfit(T2,y,1)
linear_fit = np.poly1d(coefficients)
#print(coefficients)
x_fit = np.linspace(min(T2), max(T2), 1000)
y_fit = linear_fit(x_fit)
plt.plot(x_fit,y_fit)
```

```python
g = (4*(np.pi)**2)/coefficients[0]
print(g)
sigmag= (4*(np.pi)**2)/coefficients[0]*0.01
```

```python
precision = (sigmag/g)*100
exactitud = np.abs((9.8-g))/sigmag
print(precision)
print(exactitud)
```

```python
#Desviación Estandar con eq 2.3
#Sean x los datos para usar toca adaptar a pandas
x = [25.8,26.2,26,26.5,25.8,26.1,25.8,26.3]
#Calculo del promedio
mean = np.sum(x)/len(x)

#Calculo de la desviacion   
d_i = []
for i in range(len(x)):
    d_i.append(x[i] - mean)
#print(d_i)

#Calculo de la desviacion cuadrada
d_i_squared = []
for d in d_i:
    d_i_squared.append(d ** 2)
#print(d_i_squared)

# Calculate the variance
variance = np.sum(d_i_squared) / (len(x)-1)
#print(variance)

# Calculate the standard deviation
std_dev = np.sqrt(variance)

#Calculo del Error Estandar
std_err = std_dev/np.sqrt(len(x))

print("Standard Deviation:", std_dev)
print("Standar Error:", std_err)


```

```python
# Define the standard errors
alpha_1 = 0.987654321
alpha_2 = 0.123456789

# Function to calculate lower and upper bounds
def calculate_bounds(alpha, N):
    lower_bound = alpha * (1 - 1 / np.sqrt(2 * N - 2))
    upper_bound = alpha * (1 + 1 / np.sqrt(2 * N - 2))
    return lower_bound, upper_bound

# Define the values of N
Ns = [2, 3, 5, 10, 20, 30, 100, 1000, 10000, 100000]

# Print column headers
print("{:<10s} {:<20s} {:<20s} {:<20s}".format("N", "α", "Lower Bound", "Upper Bound"))

# Calculate and print for alpha_1
print("For α =", alpha_1)
for N in Ns:
    lower_bound, upper_bound = calculate_bounds(alpha_1, N)
    print("{:<10d} {:.9f} {:.9f} {:.9f}".format(N, alpha_1, lower_bound, upper_bound))

# Calculate and print for alpha_2
print("\nFor α =", alpha_2)
for N in Ns:
    lower_bound, upper_bound = calculate_bounds(alpha_2, N)
    print("{:<10d} {:.9f} {:.9f} {:.9f}".format(N, alpha_2, lower_bound, upper_bound))
```
