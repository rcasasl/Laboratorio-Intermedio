```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as m
from linregres import linregress_w as linr
import sympy as sym
from scipy.stats import chi2
```

```python
def unpack(file_path):

    x = []
    x_title = ""
    y = []
    y_title = ""

    with open(file_path, 'r', encoding='iso-8859-1') as file:
        for line in file:
            # Process each line here
            line = line.strip()
            if line == "":
                continue
            line = line.split()
            if ord(line[0][-1]) > 57:
                x_title += line[0] + " "
                y_title += line[1] + " "
                continue
            line[0] = line[0].replace(",", ".")
            line[1] = line[1].replace(",", ".")
            x.append(float(line[0]))
            y.append(float(line[1]))
    
    return x, x_title, y, y_title
```

```python
file_path = "Actividad21-1"

x,x_title,y,y_title = unpack(file_path) 

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.xlabel(x_title)
plt.ylabel(y_title)
plt.grid(True)
plt.savefig("Gr치ficas sin an치lisis/actividad21-1.png")
```

```python
x1 = np.array([34.45,29.32,24.22,19.26])
y1 = np.array([39.04,21.11,11.64,6.49])
dy = np.array([0.01]*len(x1))
lin_reg = linr(x1,y1,dy)


f= lambda x1: lin_reg.slope()*x1+lin_reg.intercept()

x_p= np.linspace(min(x1),max(x1),500)

plt.scatter(x1,y1,c="r")
plt.grid()
plt.errorbar(x1, y1, yerr= dy, linestyle="")

plt.plot(x_p,f(x_p),c="k",linestyle="--")

print(lin_reg.slope(),"$\pm$",lin_reg.err_slope(),"\n",lin_reg.intercept(),
      "$\pm$",lin_reg.err_intercept())
```
