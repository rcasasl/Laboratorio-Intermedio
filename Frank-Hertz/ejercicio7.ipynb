```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 1.380649*(10**(-23))
r = 155*(10**(-12))

def l(T):
    P = 8.7*10**(9-(3110/T))
    n = P/(k*T)
    sigma = np.pi*(r**2)
    
    l = 1/(sigma*n*np.sqrt(2))
    return l
T1 = np.linspace(300,500)

plt.plot(T1,l(T1))
 
```
