```python
def unpack(file_path):

    x = []
    x_title = ""
    y = []
    y_title = ""

    with open(file_path, 'r', encoding='iso-8859-1') as file:
        for line in file:
            # Process each line here
            line = line.strip()
            if line == "":
                continue
            line = line.split()
            if ord(line[0][-1]) > 57:
                x_title += line[0] + " "
                y_title += line[1] + " "
                continue
            line[0] = line[0].replace(",", ".")
            line[1] = line[1].replace(",", ".")
            x.append(float(line[0]))
            y.append(float(line[1]))
    
    return x, x_title, y, y_title
```

```python
def radianes(theta):
    ang_rad = []
    for i in theta:
        rad = (i*np.pi)/180
        ang_rad.append(rad)
    return ang_rad
   
def Bragg(theta,d):
    wavelength = 2*d*np.sin(theta)
    #print(wavelength)
    return wavelength

def mu(wave,Z,k):
    muR = 1*k*(wave**3)*(Z**3)
    return muR
    
def I(mu,I_0):
    I = I_0*np.exp(-mu*d)
    return I

grad_min = np.array((20.4,20.25,20.2,20.25,20.1,20.25,20.2,20.2,20.25,20.1,20.2,20.2))

rad_min = radianes(grad_min)

I_min = np.array((0,1.0,0,2.0,2.0,2.0,4.0,3.0,6.0,6.0,9.0,10.0))

rad = radianes(xf)
k = 1
d = 2.014*(10**(-10))
Z = 29
#print(xf)
#print(rad)

x_Bragg_min = Bragg(rad_min,d)
mu1 = mu(x_Bragg_min,Z,k)
print(len(x_Bragg_min))
#print(len(mu1))
I1 = I(mu1,I_min)
```

```python
U = np.linspace(13, 35, 12)
plt.scatter(1/U, x_Bragg_min)

# Perform linear regression
coefficients = np.polyfit(1/U, x_Bragg_min, 1)
slope = coefficients[0]

print(f"Slope of the linear regression line: {slope:.4f}")

# Create a line using the fitted model
x_fit = np.linspace(1/U.min(), 1/U.max(), 100)
y_fit = np.polyval(coefficients, x_fit)

# Plot the linear regression line
plt.plot(x_fit, y_fit, color='red', label=f'Linear Fit: {coefficients[0]:.4f}x + {coefficients[1]:.4f}')

plt.xlabel('1/U')
plt.ylabel('lambda_min')
plt.legend()
plt.show()
```
